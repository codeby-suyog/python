```python
from dateutil import parser
from datetime import datetime

def parse_mixed_date(x):
    # 1) Keep NaNs as NaT
    if pd.isna(x):
        return pd.NaT
    # 2) If it's already a datetime-like, just return it
    if isinstance(x, (pd.Timestamp, datetime)):
        return x
    # 3) Otherwise try to parse the string
    try:
        return parser.parse(str(x), dayfirst=False, fuzzy=True)
    except Exception:   # catches ParserError, TypeError, ValueError, etc.
        return pd.NaT

df['Opening date'] = df['Opening date'].apply(parse_mixed_date)
```

```python
import re

# --- helpers ---
_dms_pat = re.compile(
    r'(\d+)[°º]\s*(\d+)?[′\'’]?\s*(\d+(?:\.\d+)?)?[″"\”]?\s*([NSEW])',
    flags=re.IGNORECASE
)

def _dms_to_dd(d, m=0, s=0, hemi='N'):
    d = float(d)
    m = float(m or 0)
    s = float(s or 0)
    dd = d + m/60 + s/3600
    if hemi.upper() in ('S', 'W'):
        dd = -dd
    return dd

def parse_coords(val):
    if pd.isna(val):
        return np.nan, np.nan

    s = str(val)
    # normalize spaces / weird unicode
    s = ' '.join(s.split())

    # many wiki scraps have " / " with a clean decimal pair after it:
    # e.g. "... / 29.5802°N 97.9663°W"
    tail = s.split('/')[-1] if '/' in s else s

    # 1) Try decimal degrees first (with optional N/S/E/W)
    m = re.search(
        r'([+-]?\d+(?:\.\d+)?)\s*°?\s*([NS])?.*?([+-]?\d+(?:\.\d+)?)\s*°?\s*([EW])?',
        tail,
        flags=re.IGNORECASE
    )
    if m:
        lat, lat_h, lon, lon_h = m.groups()
        lat = float(lat)
        lon = float(lon)
        if lat_h and lat_h.upper() == 'S':
            lat = -lat
        if lon_h and lon_h.upper() == 'W':
            lon = -lon
        return lat, lon

    # 2) Fall back to DMS
    parts = _dms_pat.findall(s)
    if len(parts) >= 2:
        lat_dd = _dms_to_dd(*parts[0])
        lon_dd = _dms_to_dd(*parts[1])
        return lat_dd, lon_dd

    return np.nan, np.nan

# --- apply to your df ---
df[['Latitude', 'Longititude']] = df['Coordinates'].apply(lambda x: pd.Series(parse_coords(x)))

latlon = df['Coordinates'].apply(lambda x: pd.Series(parse_coords(x), index=['Latitude', 'Longititude']))
i = df.columns.get_loc('Coordinates')

df = pd.concat(
    [df.iloc[:, :i],        # before 'Coordinates'
     latlon,                # lat & lon
     df.iloc[:, i+1:]],     # after 'Coordinates'
    axis=1
)

df = df.loc[:, ~df.columns.duplicated()]
```

```python

def extract_height_m(value):
    if pd.isna(value):
        return np.nan
    
    text = str(value).lower()
    
    # 1. Look for meters (e.g., 30.0 m)
    m_match = re.search(r'([0-9]*\.?[0-9]+)\s*m\b', text)
    if m_match:
        return round(float(m_match.group(1)), 2)
    
    # 2. Look for feet (e.g., 98.4 ft)
    ft_match = re.search(r'([0-9]*\.?[0-9]+)\s*ft\b', text)
    if ft_match:
        return round(float(ft_match.group(1)) * 0.3048, 2)
    
    return np.nan

# Apply to the column
df['Height'] = df['Height'].apply(extract_height_m)
```

```python
def extract_Length_m(value):
    if pd.isna(value):
        return np.nan
    
    text = str(value).lower()
    
    # 1. Look for meters (e.g., 30.0 m)
    m_match = re.search(r'([0-9]*\.?[0-9]+)\s*m\b', text)
    if m_match:
        return round(float(m_match.group(1)), 2)
    
    # 2. Look for feet (e.g., 98.4 ft)
    ft_match = re.search(r'([0-9]*\.?[0-9]+)\s*ft\b', text)
    if ft_match:
        return round(float(ft_match.group(1)) * 0.3048, 2)
    
    return np.nan


# Apply to the column
df['Length'] = df['Length'].apply(extract_Length_m)
```

```python
import re
import numpy as np

def extract_mph(val):
    if pd.isna(val):
        return np.nan
    text = str(val).lower()
    match = re.search(r'([0-9]*\.?[0-9]+)\s*mph', text)
    return round(float(match.group(1)), 2) if match else np.nan

df['Speed'] = df['Speed'].apply(extract_mph)
```

```python


def duration_to_minutes(val):
    if pd.isna(val):
        return np.nan
    
    text = str(val).lower().strip()
    
    # Case 1: Format "1:32" (MM:SS)
    if re.match(r'^\d+:\d+$', text):
        mins, secs = map(int, text.split(':'))
        return round(mins + secs/60, 2)
    
    # Case 2: Format "2 minutes" or "2 min"
    m = re.search(r'([0-9]*\.?[0-9]+)\s*(minute|min)', text)
    if m:
        return round(float(m.group(1)), 2)
    
    # Case 3: Only number (assume minutes)
    if text.isdigit():
        return round(float(text), 2)
    
    return np.nan

df['Duration'] = df['Duration'].apply(duration_to_minutes)
```
